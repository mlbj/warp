```python
%load_ext autoreload
%autoreload 2

import sys
sys.path.append('..')

import warp

import PIL.Image 
import numpy as np
```

### Image example

```python
image = PIL.Image.open('../data/diagram.png').convert('RGB')
image
```

```python
r,g,b=np.array(image).astype(int)[:,:,0],np.array(image).astype(int)[:,:,1],np.array(image).astype(int)[:,:,2]
r.shape
```

# warp.Butterworth

### Warp object

```python
W=warp.Butterworth(r.shape,r=100,q=50,n=5)
```

### forward

```python
butterworth_r=W.forward(r)
butterworth_g=W.forward(g)
butterworth_b=W.forward(b)
butterworth_r.shape
```

### New image

```python
data = np.zeros((r.shape[0], r.shape[1], 3), dtype=np.uint8)
data[:,:,0],data[:,:,1],data[:,:,2]=butterworth_r,butterworth_g,butterworth_b
butterworth_image = PIL.Image.fromarray(data, 'RGB')
butterworth_image
```
